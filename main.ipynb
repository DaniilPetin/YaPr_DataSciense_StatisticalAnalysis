#### Шаг 1. Загрузка данных

```python
df_users = pd.read_csv('/datasets/users_go.csv') # загрузка данных о пользователях
df_rides = pd.read_csv('/datasets/rides_go.csv') # загрузка данных о поездках
df_subs = pd.read_csv('/datasets/subscriptions_go.csv') # загрузка данных о подписках
```

```python
def check_df (df, df_name):
    print()
    print(f'\033[1mОбзор датасета {df_name}:\033[0m')
    print()

    # общая информация 
    display(df.info())

    # первые 5 строк
    display(df.head())
```

```python
check_df(df_users, 'df_users')
check_df(df_rides, 'df_rides')
check_df(df_subs, 'df_subs')
```

#### Шаг 2. Предобработка данных

##### Изменение типов данных

Обратим внимание, что столбец date датафрейма с поездками, содержащий информацию о дате поездки, имеет некорректный формат. Все остальные столбцы имеют правильный формат. Переведем данные в столбце date в формат datetime:

```python
df_rides['date'] = pd.to_datetime(df_rides['date'], format='%Y-%m-%d')
print(f"В столбце date {df_rides['date'].nunique()} уникальные даты,\
 формат данных изменился на {df_rides['date'].dtype}")
```

##### Добавление новых столбцов

Создадим новый столбец с номером месяца на основе столбца date в датафрейме с поездками:

```python
df_rides['month'] = df_rides['date'].dt.month
list = ['date', 'month']
df_rides[list].head()
```

```python
df_rides['date'].dt.year.unique()
```

Данные предоставлены только за 2021 год

```python
df_users.isna().sum() # число пропусков в датайрейме с пользователями
```

```python
print(f"Число явных дубликатов в датафрейме с пользователями равно {df_users.duplicated().sum()}")
```

Удалим явные дубликаты в этом датафрейме:

```python
df_users = df_users.drop_duplicates().reset_index(drop=True)
```

Проверим, есть ли строки, в которых содержится различающаяся информация об одном и том же пользователе:

```python
if df_users['user_id'].unique().shape[0] == df_users.shape[0]:
    print('Отсутствуют повторные упоминания одного и того же пользователя')
else:
    print('Некоторые id встречаются несколько раз')
```

Переходим к датафрейму с поездками:

```python
print(f"Число явных дубликатов в датафрейме с поездками равно {df_rides.duplicated().sum()}")
```

Повторные упоминания одной и той же поездки отсутствуют, поэтому можем оставить все данные в датасете. Искать неявные дубликаты по id пользователя или другим столбцам не имеет смысла, так как один и тот же пользователь мог совершить несколько поездок или разные поездки могли иметь одинаковые параметры.

В датасете с информацией о типах подписок в сервисе всего 2 строки. Дополнительный анализ не требуется.

##### Соотношение пользователей с подпиской и без подписки

```python
df_users['subscription_type'].value_counts()
```

```python
free_ratio = round((df_users.query('subscription_type == "free"').shape[0] / df_users.shape[0]) * 100, 1)
ultra_ratio = round((df_users.query('subscription_type == "ultra"').shape[0] / df_users.shape[0]) * 100, 1)
print(f'Число пользователей с бесплатной подпиской составляет {free_ratio}%')
print(f'Число пользователей с подпиской Ultra составляет {ultra_ratio}%')
```

```python
plt.figure(figsize=(12,5))
df_users['subscription_type'].value_counts().plot(kind = 'barh', fontsize = 16)
plt.title('Число пользователей по типам подписки', fontsize = 18)
plt.ylabel('Тип подписки', fontsize = 15)
plt.xlabel('Число пользователей', fontsize = 15)
plt.show()
```

**Вывод:** ожидаемо больше пользователей пользуются бесплатной подпиской - 835 человек, что составляет 54,4% от общего числа. Меньше людей пользуется платной подпиской Ultra - 699 человек, что соответствует 45,6% от всех представленных пользователей.

#### Шаг 6. Проверка гипотез

##### Тратят ли пользователи с подпиской больше времени на поездки?

- Нулевая гипотеза H0 - средняя продолжительность поездки для клиентов с подпиской free и с подпиской ultra равны
- Альтернативная гипотеза H1 - средняя продолжительность поездки выше у клиентов с подпиской ultra.

Для проверки гипотезы о равенстве средних двух совокурностей необходимо провести односторонний T-тест:

```python
alpha = 0.05
results = st.ttest_ind(df_ultra['duration'], df_free['duration'], 
                       equal_var = False, alternative = 'greater')
print(f'p-value теста равно {results.pvalue}')
```

```python
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не удалось отвергнуть нулевую гипотезу о равенстве продолжительности для разных подписок')
```

**Вывод:** нулевая гипотеза о равенстве продолжительности поездок для клиентов с разной подпиской отвергнута. Таким образом, есть основания предполагать, что с высокой вероятностью продолжительность поездок клиентов с подписокой выше. Поэтому такие пользователи могут быть выгоднее для компании. - Альтернативная гипотеза H1 - среднее расстояние поездок клиентов с подпиской ultra больше 3130 метров.

##### Подбор теста после обновления серверов

**Задача:** Представьте такую ситуацию: техническая команда сервиса обновила сервера, с которыми взаимодействует мобильное приложение. Она надеется, что из-за этого количество обращений в техподдержку значимо снизилось. Некоторый файл содержит для каждого пользователя данные о количестве обращений до обновления и после него. Какой тест вам понадобился бы для проверки этой гипотезы? 

**Ответ:** в данном случае потребуется провести тест гипотезы о равенстве средних для **зависимых** выборок. Выборки считаются **зависимыми**, когда данные содержат информацию об измерениях для одного и того же объекта после определенных изменений в нем. В данном случае в качестве объекта выступает пользователей сервиса, обратившихся в поддержку, а в качестве изменения - обновление серверов. Измеряемый параметр - число обращений. 
В таких случаях предлагается использовать односторонний Т-тест:

- Нулевая гипотеза H0 - число обращений в поддержку не изменилось после обновления серверов
- Альтернативная гипотеза H1 - число обращений в поддержку сократилось после обновления серверов.

Код в Python для решения поставленной задачи будет выглядеть так:

results = st.ttest_rel(after, before, alternative = 'less')

В данном случае **before** - выборка до обновления серверов, **after** - после обновления.

#### Шаг 7. Распределения

##### Акция с раздачей промокодов

**Формулировка задачи:** Отделу маркетинга GoFast поставили задачу: нужно провести акцию с раздачей промокодов на один бесплатный месяц подписки, в рамках которой как минимум 100 существующих клиентов должны продлить эту подписку. То есть по завершении периода действия подписки пользователь может либо отказаться от неё, либо продлить, совершив соответствующий платёж.

Эта акция уже проводилась ранее и по итогу выяснилось, что после бесплатного пробного периода подписку продлевают 
10% пользователей. Выясните, какое минимальное количество промокодов нужно разослать, чтобы вероятность не выполнить план была примерно 5%. Подберите параметры распределения, описывающего эту ситуацию, постройте график распределения и сформулируйте ответ на вопрос о количестве промокодов.

По завершении периода действия подписки пользователь может либо отказаться от неё, либо продлить - **независимое испытание Бернулли** с двумя элементарными исходами: **продление** подписки, **отказ** от подписки.

Соответствующее биномиальное распределение выглядит следующим образом:

```python
n = 1000
p = 0.1

distr = []

for k in range(0, n + 1):
    current_value = st.binom.pmf(k, n, p)
    distr.append(current_value)

plt.figure(figsize=(12,5))
plt.bar(range(0, n+1), distr)
plt.xlim(0, 200);
```

График симметричный, поэтому, предположительно, можем аппроксимировать биномиальное распределение нормальным распределением.

```python
n = 100
p = 0.1
k = 99
possibility = 0.05

while st.binom.cdf(k, n, p) > possibility:
    n += 1
else:
    print(f'Вероятность не выполнить план по подпискам меньше 5% при раздаче {n} промокодов')
```

```python
# Аппроксимруем биномиальное распеделение нормальным:
mu = n * p # мат. ожидание нормального распределения
sigma = (n * p * (1-p))**0.5 # стандартное отклонение нормального распределения

distr = st.norm(mu, sigma) 

# границы интервала
x = 100

# считаем вероятность получить значение от y1 до y2
result = distr.cdf(x)
print(f'Вероятность получить 100 и более продлений составляет {(1-result).round(2)*100}% и более при раздаче {n} подписок')

arange = np.arange((mu - 5 * sigma),(mu + 5 * sigma),1)

plt.figure(figsize=(12,5))
plt.plot(arange, st.norm.pdf(arange, mu, sigma))
plt.axvline(x=k, color='red', label = 'Ожидаемое число продлений')
plt.axvline(x=mu, color='green', label = 'Мат. ожидание')
plt.legend(loc='best')
plt.show()
```

##### Вероятность открытия push-уведомлений

**Формулировка задачи:** Отдел маркетинга рассылает клиентам push-уведомления в мобильном приложении. Клиенты могут открыть его или не открывать. Известно, что уведомления открывают около 40 % получивших клиентов. Отдел планирует разослать 1 млн уведомлений. С помощью аппроксимации постройте примерный график распределения и оцените вероятность того, что уведомление откроют не более 399,5 тыс. пользователей.

Данная ситуация может быть представлена биномиальным распределением с независимыми испытаниями Бернулли - клиент может открыть полученное уведомление или не открыть. Каждое испытание имеет 2 независимых исхода. Требуемое число успехов в данной задаче равно 399,5 тыс. или менее.

```python
# Параметры биномиального распределения:
n = 1000000
k = 399500
p = 0.4

mu = n * p
sigma = (n * p * (1-p))**0.5

distr = st.norm(mu, sigma) 

result = distr.cdf(k)
print(f'Вероятность открытия менее 399,5 тыс. сообщений равна {result.round(3)*100}%')

arange = np.arange((mu - 5 * sigma),(mu + 5 * sigma),1)

plt.figure(figsize=(12,5))
plt.plot(arange, st.norm.pdf(arange, mu, sigma))
plt.xticks(rotation=90)
plt.axvline(x=k, color='red', label = 'Ожидаемое число открытий')
plt.axvline(x=mu, color='green', label = 'Мат. ожидание')
plt.legend(loc='best')
plt.show()
```

**Вывод:** вероятность того, что из 1 млн. разосланных уведомлений будут открыты не более 399,5 тыс. при вероятности открытия 40%, составляет 15,4%.
